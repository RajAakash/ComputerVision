```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import random

# Assume you have a neural network with a convolutional layer
class SimpleConvNet(nn.Module):
    def __init__(self):
        super(SimpleConvNet, self).__init__()
        self.conv = nn.Conv2d(in_channels=3, out_channels=10, kernel_size=3)

    def forward(self, x):
        return self.conv(x)

# Create an instance of the neural network
net = SimpleConvNet()

# Generate a random input image (batch size of 1, 3 channels, 64x64 size)
input_image = torch.rand((1, 3, 64, 64))

# Forward pass through the neural network
image_out_of_conv = net(input_image)

# Randomly select 5 indices
random_indices = random.sample(range(0, 10), k=5)

# Create subplots
fig, axs = plt.subplots(nrows=1, ncols=5, figsize=(12, 12))

# Iterate over random indices
for i, idx in enumerate(random_indices):
    # Select the corresponding convolutional feature map
    image_conv_feature_map = image_out_of_conv[:, idx, :, :]

    # Plot the feature map on the current subplot
    axs[i].imshow(image_conv_feature_map.squeeze().detach().numpy())
    axs[i].set(xticklabels=[], yticklabels=[], xticks=[], yticks=[])

# Show the plot
plt.show()
```
